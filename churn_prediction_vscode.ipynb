# 📊 Churn Prediction (Versão Final - VS Code)

Modelo de churn com saída estruturada para negócio.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# 📂 Carregar dados
file_path = "C:/Users/Acer/Documents/PROJETOS/churn/churn/churn_silver_2025.csv"

df = pd.read_csv(file_path)
df.head()
```

```python
# 🔎 Exploração
print(df.info())
print(df.isnull().sum())
```

```python
# 🧠 Separar ID, target e features
id_cliente = df['id_cliente']
y = df['churn']

X = df.drop(['churn', 'id_cliente'], axis=1)
```

```python
# 🔄 Transformar categóricas
X = pd.get_dummies(X, drop_first=True)
```

```python
# ✂️ Split (incluindo ID)
X_train, X_test, y_train, y_test, id_train, id_test = train_test_split(
    X, y, id_cliente, test_size=0.2, random_state=42
)
```

```python
# 🤖 Treinar modelo
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# 🔮 Previsões
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]
```

```python
# 📊 Avaliação
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# 📉 Matriz de confusão
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.title('Confusion Matrix')
plt.show()
```

```python
# 📈 Distribuição do churn
sns.countplot(x=y)
plt.title('Distribuição de Churn')
plt.show()
```

```python
# 💾 Resultado final no formato esperado
df_result = pd.DataFrame({
    'id_cliente': id_test.values,
    'probabilidade_churn': y_proba,
    'previsao_final': y_pred
})

output_path = "C:/Users/Acer/Documents/PROJETOS/churn/churn/churn_resultado.csv"
df_result.to_csv(output_path, index=False)

df_result.head()
```
